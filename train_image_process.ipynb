# In this notebook, we preprocess the training images


```python
from skimage.io import imread, imsave
from skimage import color, filters, measure, util, transform, img_as_ubyte
import matplotlib.pyplot as plt
import glob
import numpy as np
import pandas as pd
%matplotlib inline
```

## glob the training images and associated masks

```python
#glob all of the training images
train_img_list = sorted(glob.glob("C:/Users/Wei/comp_540/project/stage1_train/*/images/*"))
print(len(train_img_list))
```

#Here we read in the images, and save a grayscale copy. We also
save a copy with the dark/light colors reversed.

```python
import os
from scipy import ndimage

image_dir = "C:/Users/Wei/comp_540/project/proc_train_img/"
if not os.path.exists(image_dir):
    os.makedirs(image_dir)
    
Hout = 128
Wout = 128

#demonstrate the concept with two images that are opposite dark/light
images = [ 0, 2]
for i in range(len(train_img_list)):
    img_index = i
    img_path  = train_img_list[img_index]
 #   print(img_path)
    image_id = img_path.split("\\")[-1].split('.')[0] #extract image id from path
    print("ID = ", image_id)
    img_original =imread(img_path)
    img = color.rgb2gray(img_original)#read as black and white for simple thresholding segmentation
    img /= 255 #normalize image intensities
    inverted_img = util.invert(img)
    #place the modified images in an output directory using the same format as original input
    img_out_dir = image_dir + image_id + "/images/"
    print("imoutdir  =", img_out_dir)
    mask_out_dir = image_dir + image_id + "/masks/"
    #Save the resized images and masks
    if not os.path.exists(img_out_dir):
        os.makedirs(img_out_dir)
    if not os.path.exists(mask_out_dir):
        os.makedirs(mask_out_dir)
    inv_image_id = image_id + "inv"
    inv_img_out_dir = image_dir + inv_image_id + "/images/"
    inv_mask_out_dir = image_dir + inv_image_id + "/masks/"
    #Save the inverted images and masks (the mask is unchanged, but its easier to save it separately with the inverted image)
    if not os.path.exists(inv_img_out_dir):
        os.makedirs(inv_img_out_dir)
    if not os.path.exists(inv_mask_out_dir):
        os.makedirs(inv_mask_out_dir)
    
    masks = "C:/Users/Wei/comp_540/project/stage1_train/" + image_id + "/masks/*"
    mask_list = glob.glob(masks)
    img_mask = np.zeros(img.shape,dtype='uint8')
    #build actual y-mask
    for index, mask in enumerate(mask_list):
        curr_mask = imread(mask)
        #img_mask += (curr_mask/np.amax(curr_mask) * (index + 1)).astype('uint8')
        img_mask += (curr_mask/np.amax(curr_mask)).astype('uint8') #make a binary mask for now
    #resize
    img =transform.resize(img,(Hout,Wout),preserve_range=True)
    print(img.dtype)
#    stop
    inverted_img =transform.resize(inverted_img,(Hout,Wout),preserve_range=True)
    #imsave(img_out_dir + image_id + ".png",img)
    print((img_out_dir + image_id + ".png"))
    imsave(img_out_dir + image_id + ".png",img)
    #img_mask =transform.resize(img_mask,(Hout,Wout)) 
    imsave(inv_img_out_dir + image_id + ".png",inverted_img)
    img_mask = transform.resize(img_mask,(Hout,Wout),preserve_range=True).astype('uint8')
    imsave(mask_out_dir + image_id + ".png",img_mask)
    imsave(inv_mask_out_dir + image_id + ".png",img_mask)
    #img_mask =transform.resize(img_mask,(Hout,Wout)) 
    #test = imread(img_out_dir + image_id + ".png")
    
    #Print a couple images
    if i == 0 or i == 2:
      plt.imshow(img,cmap='gray')
      title = "Original Grayscale Image  " + str(image_id)
      plt.title(title)
      plt.show()
    
      plt.imshow(inverted_img,cmap='gray')
      title = "Inverted Grayscale Image  " + str(image_id)
      plt.title(title)
      plt.show()
    
      plt.imshow(img_mask)
      title = "True Mask  " + str(image_id)
      plt.title(title)
      plt.show()
```
